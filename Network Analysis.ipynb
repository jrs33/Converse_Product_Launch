```python
import pandas as pd
import networkx as nx
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import json
```

### Chuck II

```python
chuck2_network_df = pd.read_csv("chuck2_network_df_v2.csv", encoding = "ISO-8859-1")
```

```python
chuck_graph = nx.from_pandas_dataframe(chuck2_network_df,
                                 source = 'followed_by',
                                 target = 'being_followed',
                                 edge_attr = ['followers', 'num_favorites', 'num_lists'])
```

```python
%matplotlib inline
nx.draw(chuck_graph, alpha = 0.5, node_size = 10)
plt.show()
```
